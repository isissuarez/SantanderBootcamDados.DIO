# Categorização dos municípios dos estado do Rio grande do Sul

# Contexto
 A partir de uma planilha Excel contendo os munícipios do estado do Rio grande du Sul e seu número de habitantes de acordo com o senso do ano de 2022, pretendemos identificar a quuantidade total de municípios do estado. e categoriza-los em grande , médio e pequeno.


De acordo com a categorização do IBGE (instituto brasileiro de Geografia e Estatística), em relação a quantidade de habitantes às cidades são dividídas em:

1. Cidade pequena: é aquela que possui até 99 mil habitantes.
2. Cidade média: entre 100 mil e 499 mil habitantes.
3. Cidade grande: acima de 500 mil habitantes.

Pretendemos realizar a contagem de quantos munícípios do estado podem ser encaixados em cada caregoria, bem como a quantidade total de munocípios do estado

## Entrada:
Ler uma planilha excel contendo os municípios do estado e sua população.
fonte dos dados: https://www.ibge.gov.br/estatisticas/sociais/trabalho/22827-censo-demografico-2022.html?edicao=37225&t=resultados
## Transformação
através de um algoritímo phython estabelecer as categorias dos municípios
## Load
gerar uma nova planilha excel contendo a coluna de classificação dos municípios

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('municipiosRS.xlsx')
df
```

```python
def classificacao(pop):
  if pop < 100000:
    return 'pequena'
  elif 100000 <= pop < 500000:
    return 'média'
  elif pop > 500000:
    return 'grande'
  else:
    return ' '

df['categoria'] = df['habitantes'].apply(classificacao)

df.to_excel('municipiosRS_categorizados.xlsx', index=False)
```
