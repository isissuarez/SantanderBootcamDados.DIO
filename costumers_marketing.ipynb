```python
df = pd.read_csv('Mall_Customers.csv')

df = df[['CustomerID','Age', 'Gender']]



```

```python
from openai.api_resources import completion
import openai

openai.api_key = api_key
def generate_ai_news(row):
    customer_id = row['CustomerID']
    age = row['Age']
    gender = row['Gender']

    completion = openai.ChatCompletion.create(
        model="gpt-3.5-turbo-16k",
        messages=[
            {"role": "system", "content": "Você é um especialista em marketing "},
            {"role": "user", "content": f"Crie uma mensagem para o cliente {customer_id} sobre uma loja de roupas considerando idade {age} e gênero {gender} contendo no máximo 20 caracteres"}
        ]
    )

    responseCHATGPT = completion.choices[0].message.content
    return responseCHATGPT
```

```python
df['news'] = df.apply(generate_ai_news, axis=1)
```
