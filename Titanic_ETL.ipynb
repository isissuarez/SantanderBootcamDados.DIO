# Projeto Titanic - ETL

## 1. Entrada:
Um arquivo em CSV contendo os dados dos passageiros do titanic
## 2. Transformação
Filtrar apenas os passageiros de primeira classe do sexo femino e que renham sobrevivido
## 3. Saída
Um novo aqrquivo CSV contendo as colunas Nome ( em ordem afabética), idade, sexo , classe e sobrevivente com os devidos filtros aplicados

```python
import pandas as pd

df = pd.read_csv('titanic.csv')
df

```

```python
filtro = df.loc[(df['Sex'] == 'female') & (df['Pclass'] == 1) & (df['Survived'] == 1)]
filtro2 = filtro[['Name', 'Age' ,'Sex', 'Pclass', 'Survived']]
filtroFinal = filtro2.sort_values(['Name'], ascending=True)

filtroFinal
```

```python
filtroFinal.to_csv('Titanic Solucao.csv')
```
